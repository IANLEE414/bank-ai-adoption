```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from io import StringIO
import pandas as pd
from bs4 import BeautifulSoup

POST_URL = "https://portal.kfb.or.kr/minwon/announce_member_result_test.php"

def clean_table(df: pd.DataFrame) -> pd.DataFrame:
    if isinstance(df.columns, pd.MultiIndex):
        cols = []
        for tup in df.columns:
            parts = []
            for x in tup:
                x = str(x).strip()
                if x and x.lower() != "nan" and "unnamed" not in x.lower():
                    parts.append(x)
            cols.append(" ".join(parts) if parts else "")
        df = df.copy()
        df.columns = cols
    df.columns = [c if c else f"col_{i}" for i, c in enumerate(df.columns)]
    df = df.dropna(axis=1, how="all")
    df.columns = [c.strip() for c in df.columns]
    return df

def post_via_form(payload: dict, headless=False) -> str:
    options = Options()
    options.page_load_strategy = "eager"
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--window-size=1400,900")
    options.add_argument("--blink-settings=imagesEnabled=false")
    if headless:
        options.add_argument("--headless=new")

    driver = webdriver.Chrome(options=options)
    driver.set_page_load_timeout(60)

    try:
        driver.get("about:blank")

        # form으로 POST 전송해서 결과 페이지로 이동
        js = """
        const url = arguments[0];
        const payload = arguments[1];

        const form = document.createElement('form');
        form.method = 'POST';
        form.action = url;

        for (const [k, v] of Object.entries(payload)) {
          const input = document.createElement('input');
          input.type = 'hidden';
          input.name = k;
          input.value = v;
          form.appendChild(input);
        }

        document.body.appendChild(form);
        form.submit();
        """
        driver.execute_script(js, POST_URL, payload)

        wait = WebDriverWait(driver, 30)
        # 결과 페이지에 확실히 있는 요소 기다리기: 너가 봤던 class 기준
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".search_minwon_txt")))

        return driver.page_source

    finally:
        driver.quit()

payload = {"standard": "20254", "BankCode": "11", "Gubun": "1"}
html = post_via_form(payload, headless=False)

print("has table?", "<table" in html.lower())

tables = pd.read_html(StringIO(html))
tables = [clean_table(t) for t in tables]

soup = BeautifulSoup(html, "html.parser")
meta = [li.get_text(" ", strip=True) for li in soup.select(".search_minwon_txt li.date")]

print("meta:", meta)
print("tables:", len(tables))
print(tables[0].head())

out_path = "kfb_minwon_standard20254_bank11_clean.xlsx"
with pd.ExcelWriter(out_path, engine="openpyxl") as writer:
    pd.DataFrame({"meta": meta}).to_excel(writer, index=False, sheet_name="meta")
    for i, t in enumerate(tables, start=1):
        t.to_excel(writer, index=False, sheet_name=f"table_{i}")

print("saved:", out_path)
```

```python
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

def get_meta_bankname(html: str):
    soup = BeautifulSoup(html, "html.parser")
    li = soup.select_one(".search_minwon_txt li.date")
    if not li:
        return None
    text = li.get_text(" ", strip=True)
    return text.split(":", 1)[-1].strip() if ":" in text else text.strip()

def post_via_form_fast(driver, payload: dict, wait_sec=1):
    driver.get("about:blank")
    js = """
    const url = arguments[0];
    const payload = arguments[1];
    const form = document.createElement('form');
    form.method = 'POST';
    form.action = url;
    for (const [k, v] of Object.entries(payload)) {
      const input = document.createElement('input');
      input.type = 'hidden';
      input.name = k;
      input.value = v;
      form.appendChild(input);
    }
    document.body.appendChild(form);
    form.submit();
    """
    driver.execute_script(js, POST_URL, payload)
    try:
        WebDriverWait(driver, wait_sec).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".search_minwon_txt"))
        )
        return driver.page_source
    except TimeoutException:
        return None

driver = make_driver(headless=False)
valid = []  # (code, bankname)
try:
    for code in range(1, 301):
        html = post_via_form_fast(driver, {"standard":"20254","BankCode":str(code),"Gubun":"1"}, wait_sec=1)
        if html is None:
            continue
        bankname = get_meta_bankname(html)
        if bankname:
            valid.append((code, bankname))
            print("VALID:", code, bankname)
        time.sleep(0.03)
finally:
    driver.quit()

print("\n총 유효코드 개수:", len(valid))
print("은행명 목록:", sorted(set([name for _, name in valid])))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from io import StringIO
import pandas as pd
from bs4 import BeautifulSoup
import time

# ===== URL =====
MEMBER_POST_URL = "https://portal.kfb.or.kr/minwon/announce_member_result_test.php"
CATEGORY_PAGE_URL = "https://portal.kfb.or.kr/minwon/announce_category.php"  # 안정 루트

CODES = {
    "KB국민은행": "04",
    "신한은행": "25",
    "우리은행": "20",
    "하나은행": "81",
    "카카오뱅크": "90",
    "토스뱅크": "92",
    "케이뱅크": "89",
}

def standards_2016q1_to_2024q4():
    out = []
    for year in range(2016, 2025):  # 2024 포함하려면 2025까지
        for q in range(1, 5):
            out.append(f"{year}{q}")
    return [s for s in out if 20161 <= int(s) <= 20244]

def make_driver(headless=False):
    options = Options()
    options.page_load_strategy = "eager"
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--window-size=1400,900")
    options.add_argument("--blink-settings=imagesEnabled=false")
    if headless:
        options.add_argument("--headless=new")
    driver = webdriver.Chrome(options=options)
    driver.set_page_load_timeout(60)
    return driver

def post_via_form(driver, url: str, payload: dict, wait_css: str, wait_sec=10) -> str | None:
    driver.get("about:blank")
    js = """
    const url = arguments[0];
    const payload = arguments[1];
    const form = document.createElement('form');
    form.method = 'POST';
    form.action = url;
    for (const [k, v] of Object.entries(payload)) {
      const input = document.createElement('input');
      input.type = 'hidden';
      input.name = k;
      input.value = v;
      form.appendChild(input);
    }
    document.body.appendChild(form);
    form.submit();
    """
    driver.execute_script(js, url, payload)

    try:
        WebDriverWait(driver, wait_sec).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, wait_css))
        )
        return driver.page_source
    except TimeoutException:
        return None

def flatten_columns(df: pd.DataFrame) -> pd.DataFrame:
    if isinstance(df.columns, pd.MultiIndex):
        df = df.copy()
        df.columns = [
            " ".join([str(x).strip() for x in col if str(x).strip() and str(x).lower() != "nan" and "unnamed" not in str(x).lower()])
            for col in df.columns
        ]
    df.columns = [str(c).strip() if str(c).strip() else f"col_{i}" for i, c in enumerate(df.columns)]
    df = df.dropna(axis=1, how="all")
    return df

def extract_all_tables_with_titles(html: str):
    """
    1) html에서 모든 table을 pd.read_html로 뽑고
    2) 각 table 앞(근처)에 있는 h6 텍스트를 title로 붙이려고 시도
       (사이트 구조상 h6가 테이블 섹션 제목인 경우가 많음)
    """
    soup = BeautifulSoup(html, "html.parser")
    tables = soup.find_all("table")

    dfs = []
    for idx, table in enumerate(tables):
        # title 추정: table 바로 위로 올라가며 가장 가까운 h6 찾기
        title = None
        prev = table
        for _ in range(30):  # 너무 멀리 안 가게 제한
            prev = prev.find_previous()
            if prev is None:
                break
            if prev.name == "h6":
                title = prev.get_text(" ", strip=True)
                break

        try:
            df = pd.read_html(StringIO(str(table)))[0]
        except ValueError:
            continue

        df = flatten_columns(df)
        dfs.append((idx, title, df))

    return dfs

# ====== CATEGORY는 "페이지 들어가서 form submit" 방식이 안정적 ======
def fetch_category_page(driver, standard: str, wait_sec=10) -> str | None:
    driver.get(CATEGORY_PAGE_URL)
    time.sleep(0.6)

    # 페이지마다 form/element id가 바뀔 수 있어서, JS로 standard set + submit
    # (네가 F12로 봤듯이 standard input이 존재)
    driver.execute_script(f"""
        const el = document.getElementById('standard');
        if (el) el.value = '{standard}';
        const f = document.forms[0];
        if (f) f.submit();
    """)

    try:
        WebDriverWait(driver, wait_sec).until(
            EC.presence_of_element_located((By.TAG_NAME, "table"))
        )
        return driver.page_source
    except TimeoutException:
        return None

# ====== MEMBER(은행별)은 기존 POST 유지 ======
def fetch_member_result(driver, standard, bank_code, gubun="1", retries=3):
    for _ in range(retries):
        payload = {"standard": standard, "BankCode": bank_code, "Gubun": gubun}
        html = post_via_form(
            driver,
            url=MEMBER_POST_URL,
            payload=payload,
            wait_css=".search_minwon_txt",
            wait_sec=12,
        )
        if html:
            return html
        time.sleep(1.0)
    return None

# ================= 실행 =================
driver = make_driver(headless=False)

try:
    standards = standards_2016q1_to_2024q4()

    out_rows = []  # 길게 합칠거라 "long dump" 형태

    for standard in standards:
        print("===", standard, "===")

        # (A) CATEGORY: 유형별 + 주요 금융상품별 포함(페이지 전체 table dump)
        cat_html = fetch_category_page(driver, standard)
        if cat_html is None:
            print("CATEGORY FAIL:", standard)
        else:
            cat_tables = extract_all_tables_with_titles(cat_html)
            for table_id, title, df in cat_tables:
                df = df.copy()
                df["standard"] = standard
                df["source"] = "category"
                df["table_id"] = table_id
                df["table_title"] = title
                out_rows.append(df)

        time.sleep(0.3)

        # (B) MEMBER: 은행별 (은행 루프)
        for bank_name, bank_code in CODES.items():
            mem_html = fetch_member_result(driver, standard, bank_code, retries=3)
            if mem_html is None:
                print("MEMBER FAIL:", standard, bank_name, bank_code)
                continue

            mem_tables = extract_all_tables_with_titles(mem_html)
            for table_id, title, df in mem_tables:
                df = df.copy()
                df["standard"] = standard
                df["source"] = "member"
                df["bank_name"] = bank_name
                df["bank_code"] = bank_code
                df["table_id"] = table_id
                df["table_title"] = title
                out_rows.append(df)

            time.sleep(0.2)

    all_df = pd.concat(out_rows, ignore_index=True)

    # 저장 (엑셀은 MultiIndex 문제 해결했으니 OK)
    csv_path = "kfb_all_tables_member_and_category_2016Q1_2024Q4.csv"
    xlsx_path = "kfb_all_tables_member_and_category_2016Q1_2024Q4.xlsx"

    all_df.to_csv(csv_path, index=False, encoding="utf-8-sig")
    all_df.to_excel(xlsx_path, index=False)

    print("✅ saved:", csv_path)
    print("✅ saved:", xlsx_path)

finally:
    driver.quit()
```

```python
banks = [
    '신한은행',
    '우리은행',
    '하나은행',
    '케이뱅크',
    '카카오뱅크',
    '토스뱅크'
]

# copy 중요
df2 = df[df['구분 구분'].isin(banks)].copy()

# 숫자 변환
df2['민원건수 금분기'] = pd.to_numeric(
    df2['민원건수 금분기'],
    errors='coerce'
)

# 연도 생성
df2['year'] = df2['standard'].astype(str).str[:4].astype(int)

# 은행×연도 합계
df_year_bank = (
    df2.groupby(['구분 구분','year'])['민원건수 금분기']
    .sum()
    .reset_index()
)
```
